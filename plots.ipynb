```python
import matplotlib.pyplot as plt
import numpy as np
```

Image 3: Cameraman

```python
Image= 'Cameraman'

# Noise levels
sigma = np.array([5, 10, 20, 30, 40, 50])

# SPGL_BPDN PSNR/SSIM
spgl_bpdn_psnr = np.array([31.78, 28.49, 22.94, 19.41, 16.91, 15.27])
spgl_bpdn_ssim = np.array([0.595, 0.466, 0.347, 0.276, 0.228, 0.191])

# SPGL_LASSO PSNR/SSIM
spgl_lasso_psnr = np.array([33.21, 29.16, 23.58, 20.04, 17.45, 15.73])
spgl_lasso_ssim = np.array([0.608, 0.475, 0.352, 0.278, 0.227, 0.195])

# ELASTIC_Net PSNR/SSIM
elastic_net_psnr = np.array([32.24, 28.90, 23.05, 19.59, 17.14, 15.19])
elastic_net_ssim = np.array([0.591, 0.473, 0.347, 0.277, 0.229, 0.191])

# Create the plot
plt.figure(figsize=(12, 6))

# Plot PSNR for all methods
plt.subplot(1, 2, 1)
plt.plot(sigma, spgl_bpdn_psnr, label='SPGL_BPDN PSNR', marker='o')
plt.plot(sigma, spgl_lasso_psnr, label='SPGL_LASSO PSNR', marker='o')
plt.plot(sigma, elastic_net_psnr, label='ELASTIC_Net PSNR', marker='o')
plt.xlabel('Standard deviation (, in pixels) for Image in [0,255]')
plt.ylabel('PSNR')
plt.title('PSNR vs Noise Level')
plt.legend()
plt.grid(True)

# Plot SSIM for all methods
plt.subplot(1, 2, 2)
plt.plot(sigma, spgl_bpdn_ssim, label='SPGL_BPDN SSIM', marker='o')
plt.plot(sigma, spgl_lasso_ssim, label='SPGL_LASSO SSIM', marker='o')
plt.plot(sigma, elastic_net_ssim, label='ELASTIC_Net SSIM', marker='o')
plt.xlabel('Standard deviation (, in pixels) for Image in [0,255]')
plt.ylabel('SSIM')
plt.title('SSIM vs Noise Level: ')
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
